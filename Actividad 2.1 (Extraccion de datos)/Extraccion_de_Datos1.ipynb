```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#carga desde un archivo 
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx') 
df.info()
```

```python
#Filtro por objetos
#Variables filtradas: nombre del vendedor =Leticia Hernandez
filtro1=df[(df["NOMBRE_VENDEDOR"]=="LETICIA RAMIREZ HERNANDEZ")]
filtro1
```

```python
filtro1.to_csv("Filtro1_Objetos.csv")
```

```python
#Filtro por filas
#Variables filtradas: Fila 2,5 y7
filtro2= df.iloc[[2,5,7] , :]      #Filas no consecutivas
filtro2
```

```python
filtro2.to_csv("Filtro2_Filas.csv")
```

```python
#Filtro por columnas
#Variables filtradas: Columnas 1,4 y 6
filtro3 = df.iloc[ : , [1,4,6]] 
filtro3
```

```python
filtro3.to_csv("Filtro3_Columnas.csv")
```

```python
#Se asigna la columna como indice/ creacion de index
df1=pd.read_excel('Detalle precios y productos fabricados 2022.xlsx' , index_col=2)
df1
```

```python
#Paso 2: Aplicamos filtros de filas vs columnas
#Variables filtradas: filas--> Alfredo Canela 
                     #Columnas --> cat y precio_unitario
filtro4=df1.loc[["ALFREDO CANELA"],["CANT","PRECIO_UNITARIO"]]
filtro4
```

```python
filtro4.to_csv("Filtro4_FilasvsColumnas.csv")
```

```python
#FILTRO POR CABECERA (primeros 6 datos)
filtro5= df.head(6)
filtro5
```

```python
filtro5.to_csv("Filtro5_Primeros.csv")
```

```python
#FILTRO COLA (ultimos 5 datos)
filtro6= df.tail()
filtro6
```

```python
filtro6.to_csv("Filtro6_Ultimos.csv")
```

```python
#FILTRO POR COMPARACIÓN
#Variables filtradas: precios unitarios menores a 500 pesos
filtro7= df[df["PRECIO_UNITARIO"]<500]
filtro7
```

```python
filtro7.to_csv("Filtro7_Comparacion.csv")
```

```python
#FILTRO Y
#Variables filtradas: cantidad menor a 5 y precio unitario menor a 1000
filtro8= df[(df["CANT"]<5) & (df["PRECIO_UNITARIO"]<1000)]
filtro8
```

```python
filtro8.to_csv("Filtro8_Conjuncion.csv")
```

```python
#FILTRO O
#Variables filtradas: cantidad menor a 5 o precio unitario menor a 1000
filtro9= df[(df["CANT"]<5) | (df["PRECIO_UNITARIO"]<1000)]
filtro9
```

```python
filtro9.to_csv("Filtro9_Disyuncion.csv")
```

```python
#FILTRO not
#Variables filtradas: Exclusion de las variable CANT=1
filtro10= df[~(df["CANT"]==1)]
filtro10
```

```python
filtro10.to_csv("Filtro10_Exclusion.csv")
```
