```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#carga desde un archivo 
df = pd.read_csv('Copehnague.csv')
df
```

```python
#Filtrado de 5 nombres distintos de host_name
ejercicio1 = df[df["host_name"].isin(['Dici', 'Fernando', 'Andrea', 'Carlos', 'Pato'])]
ejercicio1 
```

```python
#Filtrado de host_since incritos antes del año 2019 fue necesario hacer una verificación  de tipo detatime
df['host_since'] = pd.to_datetime(df['host_since'], errors='coerce')
ejercicio2 = df[df['host_since'] < pd.to_datetime('2019-01-01')]
ejercicio2
```

```python
#Filtro de objeto para que filtre los host que reponden en "menos de un dia"
ejercicio3=df[(df["host_response_time"]=="within a day")]
ejercicio3
```

```python
#Filtro de comparación para casas o departamentos habitables para 5 o más personas
ejercicio4=df[df["accommodates"]>= 5]
ejercicio4
```

```python
#room_type solo admite una respuesta y puede ser "Entire home/apt" o "Private room" pero no permite ambas por lo que se cambio el condicional & por |
#ejercicio5=df[(df["room_type"] == "Entire home/apt") & (df["room_type"] == "Private room")]
ejercicio5=df[(df["room_type"] == "Entire home/apt") | (df["room_type"] == "Private room")]
ejercicio5
```

```python
#Filtro de comparación para casas/departamentos con 4 camas o menos
ejercicio6=df[(df["beds"]) <=4 ]
ejercicio6
```

```python
# Filtro de conjunción y disyunción para que se filtren aquellos que tienen un baño o son super host pero que obligatoriamente sean instant_bookable
ejercicio7=df[(df["bathrooms"] == 1) | (df["host_is_superhost"] == "t") & (df["instant_bookable"] == "t")]
ejercicio7
```

```python
#Filtro de comparación y conjunción para que review_scores_cleanliness sea menor a 3 y review_scores_rating sea mayor a 3
ejercicio8=df[(df["review_scores_cleanliness"] <3 ) & (df["review_scores_rating"] >3)]
ejercicio8
```
